### Clustering K-Means

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
data_analisis = df_subset.copy()
data_analisis['Rasio_Cerai'] = (data_analisis['Jumlah Cerai'] / data_analisis['Nikah']) * 100
data_analisis['Rasio_Cerai'] = data_analisis['Rasio_Cerai'].replace([np.inf, -np.inf], 0)
```

```python
def run_kmeans(df_to_cluster, col_x, col_y):
    scaler = StandardScaler()
    X = scaler.fit_transform(df_to_cluster[[col_x, col_y]])

    kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X)
    return labels

data_analisis['Cluster_Pengangguran'] = run_kmeans(data_analisis, 'Pengangguran', 'Rasio_Cerai')
data_analisis['Cluster_Miskin'] = run_kmeans(data_analisis, 'Penduduk Miskin', 'Rasio_Cerai')

def urutkan_cluster(df_to_sort, cluster_col):
    rata2 = df_to_sort.groupby(cluster_col)['Rasio_Cerai'].mean().sort_values()
    mapping = {old: new for new, old in enumerate(rata2.index)}
    return df_to_sort[cluster_col].map(mapping)

data_analisis['Cluster_Pengangguran'] = urutkan_cluster(data_analisis, 'Cluster_Pengangguran')
data_analisis['Cluster_Miskin'] = urutkan_cluster(data_analisis, 'Cluster_Miskin')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(18, 8))

sns.scatterplot(data=data_analisis, x='Pengangguran', y='Rasio_Cerai',
                hue='Cluster_Pengangguran', palette='viridis', s=150, ax=axes[0], style='Cluster_Pengangguran')
axes[0].set_title('A. Pengaruh Pengangguran thd Perceraian', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Tingkat Pengangguran (%)')
axes[0].set_ylabel('Rasio Perceraian (%)')

sns.scatterplot(data=data_analisis, x='Penduduk Miskin', y='Rasio_Cerai',
                hue='Cluster_Miskin', palette='magma', s=150, ax=axes[1], style='Cluster_Miskin')
axes[1].set_title('B. Pengaruh Kemiskinan thd Perceraian', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Penduduk Miskin (Ribu Jiwa)')
axes[1].set_ylabel('Rasio Perceraian (%)')

for i in range(len(data_analisis)):
    if data_analisis['Cluster_Miskin'][i] >= 1:
        axes[1].text(data_analisis['Penduduk Miskin'][i], data_analisis['Rasio_Cerai'][i]+0.5,
                     data_analisis['Provinsi'][i], fontsize=9)

plt.tight_layout()
plt.show()
```

```python
print("\n" + "="*60)
print("   HASIL ANALISIS: PROVINSI PALING TERDAMPAK   ")
print("="*60)

rawan_pengangguran = set(data_analisis[data_analisis['Cluster_Pengangguran'] == 2]['Provinsi'])
rawan_miskin = set(data_analisis[data_analisis['Cluster_Miskin'] == 2]['Provinsi'])

irisan_parah = rawan_pengangguran.intersection(rawan_miskin)

print("\n1. Provinsi dengan Perceraian Tinggi yg Dipicu PENGANGGURAN (Cluster 2):")
print(f"   {', '.join(rawan_pengangguran)}")
print("\n2. Provinsi dengan Perceraian Tinggi yg Dipicu KEMISKINAN (Cluster 2):")
print(f"   {', '.join(rawan_miskin)}")

print("\n3. KESIMPULAN AKHIR (Intersection):")
if irisan_parah:
    print(f"   Provinsi yang masuk kategori 'Sangat Rawan' di kedua faktor:")
    print(f"   {', '.join(irisan_parah)}")
else:
    print(" Tidak ada provinsi yang masuk kategori 'Sangat Rawan' di kedua faktor sekaligus.")
    print(" (Artinya: Tiap provinsi punya masalah spesifik, ada yang karena miskin saja, atau pengangguran saja).")
```
